```python
import json
import matplotlib.pyplot as plt

# Load the JSON file (adjust the path if needed)
with open('/home/toukapy/Dokumentuak/Master CV/C5/mcv-c5-team2/week2/Mask2Former/mask2former_kitti_finetuned_color_v2_good_loss/checkpoint-1000/trainer_state.json', 'r') as f:
    data = json.load(f)

log_history = data["log_history"]

# Extract training and evaluation loss values along with their corresponding steps
train_steps, train_losses = [], []
eval_steps, eval_losses = [], []

for entry in log_history:
    if "loss" in entry:
        train_steps.append(entry["step"])
        train_losses.append(entry["loss"])
    if "eval_loss" in entry:
        eval_steps.append(entry["step"])
        eval_losses.append(entry["eval_loss"])

# Create the loss curve plot
plt.figure(figsize=(10, 6))
plt.plot(train_steps, train_losses, label="Train Loss", marker='o')
plt.plot(eval_steps, eval_losses, label="Eval Loss", marker='x')
plt.xlabel("Training Step")
plt.ylabel("Loss")
plt.title("Loss Curve Color Jitter")
plt.legend()
plt.grid(True)
plt.show()
```
